```python
sql = """
with ranked_data as (
select  
    device_id,
    temperature,
    timestamp,
    row_number() over (partition by device_id order by temperature desc) as row_num
from 
    cleaned_environment)
select
    device_id,
    timestamp,
    temperature as max_temperature
from 
    ranked_data
where
     row_num=1
;
"""
df = pd.read_sql_query(sql, con = engine.connect())
df
```

```python
sql = """
select  
    device_id,
    min(temperature),
    max(temperature)
from 
    cleaned_environment 
group by
    device_id
having
    min(temperature) < max(temperature)
;
"""
df = pd.read_sql_query(sql, con = engine.connect())
df
```

```python
sql = """
WITH ema_calculation AS (
    SELECT 
        device_id, 
        timestamp, 
        temperature, 
        ROW_NUMBER() OVER (PARTITION BY device_id ORDER BY timestamp) AS row_num
    FROM 
        cleaned_environment
),
recursive_ema AS (
    SELECT 
        device_id,
        timestamp,
        temperature AS ema,
        row_num
    FROM 
        ema_calculation
    WHERE 
        row_num = 1
    UNION ALL
    SELECT 
        ec.device_id,
        ec.timestamp,
        (ec.temperature * alpha) + (re.ema * (1 - alpha)) AS ema,
        ec.row_num
    FROM 
        ema_calculation ec
    JOIN 
        recursive_ema re
    ON 
        ec.device_id = re.device_id AND ec.row_num = re.row_num + 1
)
SELECT 
    device_id, 
    timestamp, 
    ema
FROM 
    recursive_ema
ORDER BY 
    device_id, timestamp
LIMIT 10;

;
"""
df = pd.read_sql_query(sql, con = engine.connect())
df
```

```python

df1 = pd.read_csv("cleaned_environment.csv")
df1['timestamp'] = pd.to_datetime(df1['timestamp'])

df1 = df1.sort_values(by=['device_id', 'timestamp'])

alpha = 0.1

# Compute EMA for each device
df1['ema'] = df1.groupby('device_id')['temperature'].apply(lambda x: x.ewm(alpha=alpha, adjust=False).mean())

# Display top 10 rows
print(df1.head(10))
```

```python
sql = """
SELECT 
    device_id, 
    MAX(humidity) AS max_humidity
FROM 
    cleaned_environment
GROUP BY 
    device_id;
"""
df = pd.read_sql_query(sql, con = engine.connect())
df
```

```python
sql = """
WITH temperature_stats AS (
    SELECT 
        device_id, 
        AVG(temperature) AS mean_temperature, 
        STDDEV(temperature) AS std_dev_temperature
    FROM 
        cleaned_environment
    GROUP BY 
        device_id
)
SELECT 
    dr.device_id, 
    AVG(dr.temperature) AS average_temperature_within_3_std_dev
FROM 
    cleaned_environment dr
JOIN 
    temperature_stats ts
ON 
    dr.device_id = ts.device_id
WHERE 
    dr.temperature BETWEEN (ts.mean_temperature - 3 * ts.std_dev_temperature) 
                        AND (ts.mean_temperature + 3 * ts.std_dev_temperature)
GROUP BY 
    dr.device_id;
"""
df = pd.read_sql_query(sql, con = engine.connect())
df
```

```python
sql = """
SELECT 
    device_id,
    CASE 
        WHEN WEEKDAY(timestamp) < 5 THEN 'Weekday'
        ELSE 'Weekend'
    END AS day_type,
    AVG(temperature) AS average_temperature
FROM 
    cleaned_environment
GROUP BY 
    device_id, 
    day_type
ORDER BY 
    device_id, 
    day_type;
"""
df = pd.read_sql_query(sql, con = engine.connect())
df
```

```python
sql = """
SELECT 
    device_id, 
    timestamp, 
    temperature,
    SUM(temperature) OVER (PARTITION BY device_id ORDER BY timestamp) AS cumulative_temperature
FROM 
    cleaned_environment
ORDER BY 
    device_id, 
    timestamp
LIMIT 10;
"""
df = pd.read_sql_query(sql, con = engine.connect())
df
```
